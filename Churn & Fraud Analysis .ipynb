```python
## FILLING IN THE MISSING VALUES
df.fillna({'TransactionID': 't7788377377373', 
           'UserID': 'U_7455', 
           'TransactionDate': '2024-06-05',
           'TransactionTime' : '12:00',
           'Amount' : '500',
           'Location' : 'lagos',
           'IsVerified' : 'True',
           'AccountTier': 'Tier1',
           'Status' : 'Success',
           'Channel' : 'web',
           'DeviceType' : 'ios',
           'RecipientID' : 'u055'}, inplace=True)
```

```python
## filling in the missing value
df.fillna ({'UserType': 'personal'}, inplace=True)
```

```python
## checking for missing values

df.isnull().sum()
```

```python
## fixin the amount colum
df['Amount'] =  (
                 df['Amount']
                 .str.replace('$', '', regex=False)
                 .str.replace(',', '', regex=False)
                 .str.replace('NGN', '', regex=False)
                 .astype(float))
                                       
```

```python
df['Amount']
```

```python
# Clean and convert transaction_date
df['TransactionDate'] = df['TransactionDate'].astype(str).str.strip()
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], errors='coerce')
```

```python
## checking for the data type
df.dtypes
```

```python
# Scan for suspicious characters in all object columns
import re

text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    sample_vals = df[col].dropna().astype(str).head(1500).tolist()
    print(f"\nColumn: {col}")
    for val in sample_vals:
        if re.search(r'[\n\r\t\xa0]', val):
            print(f" Found suspicious characterin:{val}")
```

```python
## create last transaction date
last_transaction = df.groupby('UserID')['TransactionDate'].max().reset_index()
last_transaction.columns = ['UserID', 'LastTransactionDate']
df = df.merge(last_transaction, on='UserID', how='left')
```

```python
### creating the days since last transaction column
reference_date = pd.to_datetime("2025-01-01")
last_transaction["DaysSinceLastTxn"] =(reference_date - df['LastTransactionDate']).dt.days

## creating the churned column

last_transaction['churned'] = last_transaction['DaysSinceLastTxn'] > 90
df = df.merge(last_transaction[['UserID', 'LastTransactionDate', 'DaysSinceLastTxn',  'churned']], on='UserID', how='left')
df
```

```python
## deleting the duplicate column
df = df.drop(columns = ['LastTransactionDate_y'])
```

```python
# Sort by user and transaction date
df = df.sort_values(by=['UserID', 'TransactionDate'])
```

```python
# Function to flag if a user was reactivated
def detect_reactivation(group):
    reactivated = False
    was_churned = False
    
    for days in group['DaysSinceLastTxn']:
        if was_churned and days <= 90:
            reactivated = True
            break
        if days > 90:
            was_churned = True
            
    return pd.Series({'IsReactivated': reactivated})

# Apply the function by user
reactivation_flags = df.groupby('UserID', group_keys=False).apply(detect_reactivation).reset_index()

# Merge this back to your original df or keep it as a user summary
df_users = df[['UserID']].drop_duplicates().merge(reactivation_flags, on='UserID', how='left')
```
